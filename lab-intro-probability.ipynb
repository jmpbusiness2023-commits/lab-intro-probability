**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
from scipy.stats import norm, expon, poisson, binom, geom
```

```python
#code here

n = 460 # 460 tickets sold
p = 0.97 # 97% chance of success (3% chance of missing a flight)

binom_dist = binom(n, p)
binom_dist.pmf(450) # Probability of exactly 450 passengers showing up
print(f"There is a {binom_dist.pmf(450) * 100:.2f}% chance that exactly 450 passengers will show up.")
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
#code here

p = 0.3 # 30% chance of resolving a customer complaint on the first call
n = 3  # 3 customer calls

geom_dist = geom(p)

# Probability of resolving a complaint after at least 3 calls
probability = geom_dist.pmf(3)
print(f"There is a {probability * 100:.2f}% chance that representatives need at least 3 attempts to resolve a customer complaint.")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
#code here

mu = 500  # average number of visits per hour
k = 550  # threshold for server being overwhelmed

# P(X > k) = 1 - P(X ≤ k)

probability = 1 - poisson.cdf(k, mu)
print(f"There is a {probability * 100:.2f}% chance that the website server will be overwhelmed.")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
#code here

# Assuming that each hour’s traffic is independent,
# we can calculate the probability of no overload in 24 hours: (1 - P_overload_hour)^24
# Then, P(at least one overload) = 1 - (1 - P_overload_hour)^24

daily_probability = 1 - (1 - probability) ** 24
print(f"There is a {daily_probability * 100:.2f}% chance that the website server will be overwhelmed at some point during a day.")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
#code here

lambda_rate = 1 / 10  # rate parameter (1/mean)
x = 5  # time interval in minutes

expon_dist = expon(scale=1/lambda_rate)
probability = expon_dist.cdf(x)

print(f"There is a {probability * 100:.2f}% chance that the next customer will arrive within the next 5 minutes.")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
#code here

y = 15  # time interval in minutes
expon_dist = expon(scale=1/lambda_rate)
probability_break = expon_dist.cdf(y)

print(f"There is a {probability_break * 100:.2f}% chance that an employee will take a break after 15 minutes without a customer.")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
#code here

mean = 150
std_dev = 10

probability = norm.cdf(160, mean, std_dev) - norm.cdf(140, mean, std_dev)
print(f"There is a {probability * 100:.2f}% chance that a randomly selected bird weighs between 140 and 160 grams.")
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
#code here

mean_lifetime = 50
lambda_rate = 1 / mean_lifetime
x = 30
expon_dist = expon(scale=1/lambda_rate)
probability_fail = expon_dist.cdf(x)

print(f"There is a {probability_fail * 100:.2f}% chance that the component will fail within the first 30 hours.")
```
